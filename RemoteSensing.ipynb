```python
conversion = 272.15
paths = glob(f'mw_ir_oi/*/*/*.nc') #folder netcdf
filenames = [path.split('/')[3][:8] for path in paths]
paths = {k:v for k,v in zip(filenames,paths)}
paths
```

```python
from datetime import timedelta, date
nama = 'Januari2020'
start_month = date(2020, 1, 1)
end_month = date(2020, 1, 31)
dates = []
for n_date in range(1,19):
  dates = pd.date_range(start_month,end_month).astype(str).str.replace('-','')
  dates = dates.append(dates)
  start_month= date(start_month.year-1,start_month.month,start_month.day)
  end_month= date(end_month.year-1,end_month.month,end_month.day)
```

```python
gambars = []
for date in dates:
    data = nc.Dataset(paths[date])
    lon = np.array(data['lon'][:])
    lat = np.array(data['lat'][:])
    gambar = np.array(data['analysed_sst'][:])[0][np.where((lat>-10)&(lat<10))][:,np.where((lon>90)&(lon<140))][:,0,:] #molluccas sea coordinate
    gambar = gambar-conversion
    gambar = cv2.flip(gambar,0)
    gambars.append(gambar)
gambars = np.array(gambars)
gambars = np.mean(gambars,axis=0)
```

```python
ayam = gambars.astype(float)
ayam[gambars<0] = np.nan
# ayam = np.isnan(gambars).sum()
```

```python
original_image = cv2.imread('data1/image_result/%s.png' %nama)
img=cv2.cvtColor(original_image,cv2.COLOR_BGR2RGB)
vectorized = img.reshape((-1,3))
vectorized = np.float32(vectorized)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

```python
K = 3
attempts=10
ret,label,center=cv2.kmeans(vectorized,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS) 
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))

plt.figure(figsize=(20,20))
plt.subplot(1,2,2)
plt.imshow(result_image, extent=(90,140,-10,10))
plt.title('Segmented Image when K = %i' % K)
plt.axis('on')
path = "./data1/image_result/aftercluster/"
plt.savefig(path + '%s'%nama +' untuk k-%i' % K, bbox_inches='tight')
plt.show()
```

```python
from sklearn.cluster import KMeans
SSE = []
K = range(1,10)
for k in K:
    kmeans = KMeans(n_clusters=k).fit(vectorized)
    #print(data["clusters"])
    SSE.append(kmeans.inertia_) # Inertia: Sum of distances of samples to their closest cluster center
plt.figure(figsize=(16,8))
plt.plot(K[2:], SSE[2:])
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.title('The Elbow Method showing the optimal k')
hasil = "./data1/Hasil image/Elbow"
plt.savefig(hasil + '%s'%nama +' elbow method')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

no_of_clusters = range(3,11)

for n_clusters in no_of_clusters:
    cluster = KMeans(n_clusters = n_clusters)
    cluster_labels = cluster.fit_predict(vectorized)
    # The silhouette_score gives the
    # average value for all the samples.
    silhouette_avg = silhouette_score(vectorized, cluster_labels)
    print("For no of clusters =", n_clusters," The average silhouette_score is :", silhouette_avg)
```
